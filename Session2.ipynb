## 2. Vector Operations

### 2.1 Vector Definition and Basic Operations


#### Mathematical Background
A vector v in n-dimensional space is an ordered list of n numbers: v = (v₁, v₂, ..., vₙ)

Basic operations:
1. Vector Addition: u + v = (u₁ + v₁, u₂ + v₂, ..., uₙ + vₙ)
2. Scalar Multiplication: For a scalar c, cv = (cv₁, cv₂, ..., cvₙ)

```python
import numpy as np

# Define vectors
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

# Vector addition
w = u + v

# Scalar multiplication
c = 2
cv = c * v

print(f"u + v = {w}")
print(f"{c} * v = {cv}")
```

### 2.2 Dot Product


#### Mathematical Background
The dot product of two vectors u and v is defined as:
u · v = u₁v₁ + u₂v₂ + ... + uₙvₙ

Properties:
- Commutative: u · v = v · u
- Distributive: u · (v + w) = u · v + u · w
- Scalar multiplication: (cu) · v = c(u · v) = u · (cv)

```python
import numpy as np

u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

dot_product = np.dot(u, v)
print(f"u · v = {dot_product}")



# Calculate the dot product of a = (2, -1, 3) and b = (0, 4, -2).


# a = np.array([2, -1, 3])
# b = np.array([0, 4, -2])

# dot_product = np.dot(a, b)
# print(f"a · b = {dot_product}")
```

### 2.3 Cross Product


#### Mathematical Background
The cross product of two 3D vectors u and v is defined as:
u × v = (u₂v₃ - u₃v₂, u₃v₁ - u₁v₃, u₁v₂ - u₂v₁)

Properties:
- Anti-commutative: u × v = -(v × u)
- Distributive: u × (v + w) = (u × v) + (u × w)
- Scalar multiplication: (cu) × v = c(u × v) = u × (cv)

```python
import numpy as np

u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

cross_product = np.cross(u, v)
print(f"u × v = {cross_product}")
```

```python
import numpy as np

# Define vectors r and s
r = np.array([2, -1, 3])
s = np.array([1, 2, -1])

# Calculate r × s
r_cross_s = np.cross(r, s)
print(f"r × s = {r_cross_s}")

# Calculate s × r
s_cross_r = np.cross(s, r)
print(f"s × r = {s_cross_r}")

# Verify the anti-commutative property
if np.array_equal(r_cross_s, -s_cross_r):
  print("The anti-commutative property is verified.")
else:
  print("The anti-commutative property is not verified.")
```

## 3. Matrix Operations

### 3.1 Matrix Definition and Basic Operations


#### Mathematical Background
A matrix is a rectangular array of numbers. An m × n matrix A has m rows and n columns:

A = [a₁₁ a₁₂ ... a₁ₙ]
    [a₂₁ a₂₂ ... a₂ₙ]
    [ ...         ]
    [aₘ₁ aₘ₂ ... aₘₙ]

Basic operations:
1. Matrix Addition: (A + B)ᵢⱼ = Aᵢⱼ + Bᵢⱼ
2. Scalar Multiplication: (cA)ᵢⱼ = cAᵢⱼ

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix addition
C = A + B

# Scalar multiplication
c = 2
cA = c * A

print(f"A + B =\n{C}")
print(f"{c} * A =\n{cA}")
```

### 3.2 Matrix Multiplication

#### Mathematical Background
For an m × n matrix A and an n × p matrix B, their product AB is an m × p matrix:

(AB)ᵢⱼ = Σₖ(Aᵢₖ * Bₖⱼ) for k = 1 to n

Properties:
- Not commutative: AB ≠ BA in general
- Associative: (AB)C = A(BC)
- Distributive: A(B + C) = AB + AC

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

C = np.dot(A, B)
print(f"AB =\n{C}")
```

## 4. Eigenvalues and Eigenvectors

### Mathematical Background
For a square matrix A, a scalar λ is an eigenvalue of A if there exists a non-zero vector v such that:

Av = λv

The vector v is called an eigenvector of A corresponding to the eigenvalue λ.

```python
import numpy as np

A = np.array([[4, -2], [1, 1]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

### Exercise 4
Calculate the eigenvalues and eigenvectors of the matrix N = [[2, 1], [1, 2]]. Verify that Nv = λv for each eigenpair (λ, v).

```python
import numpy as np

# Define the matrix N
N = np.array([[2, 1], [1, 2]])

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(N)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Verify Nv = λv for each eigenpair
for i in range(len(eigenvalues)):
  eigenvalue = eigenvalues[i]
  eigenvector = eigenvectors[:, i]

  Nv = np.dot(N, eigenvector)
  lambda_v = eigenvalue * eigenvector

  print(f"\nEigenpair {i+1}:")
  print("Nv:", Nv)
  print("λv:", lambda_v)

  # Check if Nv and λv are approximately equal
  if np.allclose(Nv, lambda_v):
    print("Nv = λv is verified for this eigenpair.")
  else:
    print("Nv = λv is not verified for this eigenpair.")
```

## 5. Basic Image Operations and Filters: Applying Linear Algebra to Digital Images

In this section, we'll explore how linear algebra concepts are applied in image processing. We'll cover basic image operations and filters, explaining the mathematical theory behind each and implementing them in Python.


### 5.1 Introduction to Digital Images

A digital image is represented as a matrix of pixel values. For a grayscale image, this is a 2D matrix, while for a color image, it's typically represented as three 2D matrices (one each for Red, Green, and Blue channels).


### 5.2 Basic Image Operations


#### 5.2.1 Image Resizing

**Theory:** Resizing an image involves creating a new matrix with different dimensions and interpolating values from the original image.


```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def resize_image(image, new_size):
    return np.array(Image.fromarray(image).resize(new_size))

img = np.array(Image.open('/content/desk.jpeg'))

# Resize the image
resized_img = resize_image(img, (20, 200))

# Display results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(img)
ax1.set_title('Original')
ax2.imshow(resized_img)
ax2.set_title('Resized')
plt.show()
```

#### 5.2.2 Image Rotation
**Theory:** Rotating an image involves applying a rotation matrix to each pixel coordinate.


```python
def rotate_image(image, angle):
    return np.array(Image.fromarray(image).rotate(angle))

# Rotate the image
rotated_img = rotate_image(img, 90)

# Display results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(img)
ax1.set_title('Original')
ax2.imshow(rotated_img)
ax2.set_title('Rotated')
plt.show()
```

### 5.3 Basic Image Filters


#### 5.3.1 Grayscale Conversion
**Theory:** Grayscale conversion is a weighted sum of the RGB channels, which is essentially a dot product operation.


```python
def to_grayscale(image):
    return np.dot(image[..., :3], [0.299, 0.587, 0.114]).astype(np.uint8)

# Apply grayscale filter
gray_img = to_grayscale(img)

# Display results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(img)
ax1.set_title('Original')
ax2.imshow(gray_img, cmap='gray')
ax2.set_title('Grayscale')
plt.show()
```

#### 5.3.2 Brightness Adjustment
**Theory:** Brightness adjustment is a simple scalar addition to the pixel values.


```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def adjust_brightness(image, factor):
    return np.clip(image * factor, 0, 255).astype(np.uint8)

# Apply brightness filter
bright_img = adjust_brightness(img, 1.5)  # Increase brightness by 50%
dark_img = adjust_brightness(img, 0.5)   # Decrease brightness by 50%

# Display results
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
ax1.imshow(img)
ax1.set_title('Original')
ax2.imshow(bright_img)
ax2.set_title('Brightened')
ax3.imshow(dark_img)
ax3.set_title('Darkened')
plt.show()
```

#### 5.3.3 Contrast Adjustment
**Theory:** Contrast adjustment involves scalar multiplication of the pixel values.


```python
def adjust_contrast(image, factor):
    return np.clip((image - 128) * factor + 128, 0, 255).astype(np.uint8)

# Apply contrast filter
high_contrast = adjust_contrast(img, 1.5)
low_contrast = adjust_contrast(img, -0.5)

# Display results
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
ax1.imshow(img)
ax1.set_title('Original')
ax2.imshow(high_contrast)
ax2.set_title('High Contrast')
ax3.imshow(low_contrast)
ax3.set_title('Low Contrast')
plt.show()
```
